```python
#Agrregate frame level data to time slice data
import pandas as pd
emotion_df = pd.read_csv("C://Users/ssoud/VideoEmotion/logs/LondonEnglishEmotionList.tsv",sep='\t')

writefilename='C://Users/ssoud/VideoEmotion/logs/LondonEnglishEmotionAggregation_5Min.tsv'
f = open(writefilename, "a")
f.write("FaceId\tCoordinates\tTimeSlotNo\tEmotion\tSentiment"+"\n")
f.close()
timeslotNo=1
framerange=25*5
noofIterations = round(emotion_df.shape[0]/framerange)+1
for index in range(1,noofIterations):
    startIndex=framerange*timeslotNo
    endIndex=startIndex+framerange
    df_1 = emotion_df.iloc[startIndex:endIndex, :]
    
    print("Start Index "+str(startIndex))
    print(df_1.shape[0])
    emotioncount = df_1['Emotion'].value_counts()
    print(emotioncount)
    emotionrank = emotioncount.rank()
    emotionrank = emotionrank.index[0]
    
#emotiontop = emotionrank.head(1)

    sentimentrank = df_1['Sentiment'].value_counts().rank().index[0]
    #print(sentimentrank)
    coordinates = df_1.iloc[-1]["Coordinates"]
    FaceId = df_1.iloc[-1]["FaceId"]

    f = open(writefilename, "a")
    f.write(str(FaceId)+"\t"+str(coordinates)+"\t"+str(timeslotNo)+"\t"+str(emotionrank)+"\t"+str(sentimentrank)+"\n")
    f.close()
    timeslotNo=timeslotNo+1
```

```python
import pandas as pd
emotion_df = pd.read_csv("C://Users/ssoud/VideoEmotion/logs/ClimateChangeEmotionList.tsv",sep='\t')

writefilename='C://Users/ssoud/VideoEmotion/logs/ClimateChangeEmotionList_5Min.tsv'
f = open(writefilename, "a")
f.write("FaceId\tCoordinates\tTimeSlotNo\tEmotion\tSentiment"+"\n")
f.close()
timeslotNo=1
framerange=25*5
noFaces=5
noofIterations = round(emotion_df.shape[0]/framerange)+1


for index in range(1,noofIterations):
    
    for face in range(1,noFaces+1,1):
        startIndex=framerange*timeslotNo
        endIndex=startIndex+framerange
        #df_1 = emotion_df.iloc[startIndex:endIndex, :]
        #Select the corresponding Face and Corresponding window
        df_1 = emotion_df.loc[(emotion_df.FaceId==face) & (emotion_df.FrameId>=startIndex)& (emotion_df.FrameId<=endIndex)]
        if(df_1.shape[0]!=0):
            emotioncount = df_1['Emotion'].value_counts()
            print(emotioncount)
            emotionrank = emotioncount.rank()
            emotionrank = emotionrank.index[0]
    
#emotiontop = emotionrank.head(1)

            sentimentrank = df_1['Sentiment'].value_counts().rank().index[0]
    #print(sentimentrank)
            coordinates = df_1.iloc[-1]["Coordinates"]
            FaceId = df_1.iloc[-1]["FaceId"]

            f = open(writefilename, "a")
            f.write(str(FaceId)+"\t"+str(coordinates)+"\t"+str(timeslotNo)+"\t"+str(emotionrank)+"\t"+str(sentimentrank)+"\n")
            f.close()
        timeslotNo=timeslotNo+1
```
